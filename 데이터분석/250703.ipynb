```python
from bs4 import BeautifulSoup
import requests
```

```python
# 실습 1. 네이버 주가 크롤링

res = requests.get("https://finance.naver.com/sise/sise_quant.nhn")
soup = BeautifulSoup(res.text, 'html.parser')
```

```python
soup.select('tr')[0]
```

```python
soup.select('td')
```

```python
table = soup.select_one('table.type_2')
rows = table.select('tr')[2:]
rows
```

```python
result = []

for row in rows:
    columns = row.select('td')
    if len(columns) > 1:
        name = columns[1].text.strip()
        price = columns[2].text.strip()
        result.append({'name': name, 'price': price})

result
```

```python
rows = soup.select('table.type_2 tr')
rows
```

```python
# 2. 전일 내비 상승한 종목

result = []

for row in rows:
    columns = row.select('td')
    if len(columns) > 1:
        change_tag = row.select_one('td .blind')
        if change_tag and change_tag.text.strip() == '상승':
            name = columns[1].text.strip()
            price = columns[2].text.strip()
            result.append({'name': name, 'price': price})

result
```
