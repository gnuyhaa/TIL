```python
from bs4 import BeautifulSoup
import requests
```

```python
# 실습 1. 네이버 주가 크롤링

res = requests.get("https://finance.naver.com/sise/sise_quant.nhn")
```

```python
res.status_code # 200
res.text.find('KODEX 인버스') 
# 정적웹
```

```python
soup = BeautifulSoup(res.text, 'html.parser')
```

```python
soup.select('tr')[-4]
```

```python
table = soup.select_one('table.type_2')
rows = table.select('tr')[2:]
rows
```

```python
# 다른 풀이 방법 
result = []

for row in rows:
    temp = {}
    # row.select_one('td:nth-child(3)')
    
    columns = row.select('td')
    columns[1].text # 종목명
    columns[2].text # 현재가
    columns[3] # 전일비

    name_tag = row.select_one('a.tltle')
    price_tag = row.select_one('td.number')
    if name_tag and price_tag:
        temp['name'] = name_tag.text
        temp['price'] = price_tag.text
        result.append(temp)
result
```

```python
result = []

for row in rows:
    columns = row.select('td')
    if len(columns) > 1:
        name = columns[1].text
        price = columns[2].text
        result.append({'name': name, 'price': price})

result
```

```python
rows = soup.select('table.type_2 tr')
rows
```

```python
# 다른 풀이 방법 
result = []

for row in rows:
    temp = {}

    name_tag = row.select_one('a.tltle')
    price_tag = row.select_one('td.number')
    change_tag = row.select_one('.bu_pup') # 전일비 상승 클래스(em)
    change_tag = row.select_one('span') # 전일비 span 태그

    if name_tag and change_tag.text == '상승':
        temp['name'] = name_tag.text
        temp['price'] = price_tag.text
        result.append(temp)
result
```

```python
# 다른 풀이 2
result = []

for row in rows:
    columns = row.select('td')
    if len(columns) > 1:
        print(columns[3].select_one('span').text == '상승')
        print('bu_pup' in columns[3].select_one('em').attrs['class'])
        print('red02' in columns[3].select_one('.tah').attrs['class'])
        # print(columns[3].attrs['class']) # 안되는 이유 -> number만 가져오기 때문

        change_tag = row.select_one('td .blind')
        if change_tag and change_tag.text == '상승':
            name = columns[1].text
            price = columns[2].text
            result.append({'name': name, 'price': price})

result
```

```python
# 2. 전일 대비 상승한 종목

result = []

for row in rows:
    columns = row.select('td')
    if len(columns) > 1:
        change_tag = row.select_one('td .blind')
        if change_tag and change_tag.text == '상승':
            name = columns[1].text
            price = columns[2].text
            result.append({'name': name, 'price': price})

result
```
