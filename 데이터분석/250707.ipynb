```python
def languages(week):
    if week == 'MONDAY':
        return '월'
    elif week == 'TUESDAY':
        return '화'
    elif week == 'WEDNESDAY':
        return '수'
    elif week == 'THURSDAY':
        return '목'
    elif week == 'FRIDAY':
        return '금'
    elif week == 'SATURDAY':
        return '토'
    else:
        return '일'
```

```python
data = res.json()
weeks = ['MONDAY', 'TUESDAY', 'WEDNESDAY','THURSDAY','FRIDAY', 'SATURDAY','SUNDAY']
data = data['titleListMap']
result = []


for week in weeks:
    comics = data[week]
    for comic in comics:
        result.append({
            '제목' : comic['titleName'],
            '링크' : f"https://comic.naver.com/webtoon/list?titleId={comic['titleId']}",
            '요일' : languages(week)
        })
result
```

### 2. 로켓펀치 채용 크롤링
#### 실습 - 로켓펀치 관련 힌트
- headers에 키-값이 더 필요함 (x-...)
- 결과물은 기업명, 채용공고명만 정리할 것
- 다음 페이지 가져올 때 pageToken이 어디에 담겨있는지 확인
- 맨 첫 번째 페이지는 pageToken이 필요한지 확인

```python
res = requests.get("https://www.rocketpunch.com/api/proxy/jobs?sort=DATE_DESC",
                   headers={
                       'user-agent':"Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/138.0.0.0 Safari/537.36",
                       'x-rocket-client-id' : "63c20c18-7a24-4b2e-994e-ebec8c6139ce",
                       'x-rocket-client-type' : "WEB",
                       'x-rocket-device-type' : 'PC',
                       'x-rocket-os-type' : 'WIN',
                       'x-rocket-api-version' : ''
                   })
res.status_code
```

```python
jobs = res.json()['items']
jobs
```

```python
# companyName 기업명
# title 채용공고명 
```

```python
result = []
for job in jobs:
    result.append({
        '기업명' : job['companyName'],
        '채용공고명' : job['title'],
    })
result
```

```python
jobs = res.json()['items']
token = res.json()['pageToken']
token
```

```python
token = ''
result = []

for i in range(10):
    res = requests.get(f"https://www.rocketpunch.com/api/proxy/jobs?sort=DATE_DESC&pageToken={token}",
                    headers={
                        'user-agent':"Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/138.0.0.0 Safari/537.36",
                        'x-rocket-client-id' : "63c20c18-7a24-4b2e-994e-ebec8c6139ce",
                        'x-rocket-client-type' : "WEB",
                        'x-rocket-device-type' : 'PC',
                        'x-rocket-os-type' : 'WIN',
                        'x-rocket-api-version' : ''
                    })
    jobs = res.json()['items']
    token = res.json()['pageToken']

    for job in jobs:
        result.append({
            '기업명' : job['companyName'],
            '채용공고명' : job['title'],
        })
result
```
